```python
from MIAPCA import PCAMIA
import numpy as np
import matplotlib.pyplot as plt
from rasterio.plot import show
```

```python
mask = np.where((pca.scores[:,1]<-2) & (pca.scores[:,2]<0), 1, 0)


score_plot = np.reshape(mask, newshape=(pca.new_rows, pca.new_columns))
    
plt.imshow(score_plot, cmap=plt.get_cmap('binary'))
#show(pca.image)
plt.colorbar()
plt.title("Dummy image for the selected features")

plt.show()
```
